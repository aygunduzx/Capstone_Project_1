```python
# Distrubution of Damage based on accident and incident
df2=pd.read_csv('out_ntsb.csv',index_col=0)
df2=df2.groupby(['Investigation Type','Aircraft Damage']).size().unstack().plot(kind='bar')
```

```python
df2=pd.read_csv('out_ntsb.csv',index_col=0)
df2=df2.groupby(['Event Date','Aircraft Damage']).size().unstack().plot(kind='bar')
```

```python
# Effect of number of engines
df2=pd.read_csv('out_ntsb.csv',index_col=0)
df2.groupby('Number of Engines').size().plot(kind='bar', legend=True)

plt.show()
```

```python
# Which phase of flight is the most critical?
df2=pd.read_csv('out_ntsb.csv',index_col=0)
df2.groupby('Broad Phase of Flight').size().plot(kind='bar')
plt.xlabel('Phase of Flight')
plt.ylabel('Count')
plt.title('Accidents')
plt.show()
```

```python
# Which weather type is the most critical, Visual Meteorological Condions or Instrument Meteorological Conditions?
df2=pd.read_csv('out_ntsb.csv',index_col=0)
df2.groupby('Weather Condition').size().plot(kind='bar')

plt.show()
```
